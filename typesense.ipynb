```python
client = typesense.Client({
    'nodes': [{
        'host': 'r3zf69osqvihyek7p-1.a2.typesense.net',    
        'port': '443',
        'protocol': 'https'
    }],
    'api_key' : 'aogRh3JTT8alE8qY7whmtZf6gASLqOZ9',
    'connection_timeout_seconds': 2
})

books_schema = {
    'name': 'books',
    'fields': [
        {'name': 'title', 'type': 'string'},
        {'name': 'author', 'type': 'string[]', 'facet': True},
        {'name': 'publication_year', 'type': 'int32', 'facet': True},
        {'name': 'ratings_count', 'type': 'int32'},
        {'name': 'average_rating', 'type': 'float'},
        {'name': 'summary', 'type': 'string'},
        {'name': 'genre', 'type': 'string[]'}
    ],
    'default_sorting_field': 'ratings_count'
}
from typesense.exceptions import ObjectAlreadyExists

try:
    collection = client.collections.create(books_schema)
    print("Collection created:", collection)
except ObjectAlreadyExists:
    collection = client.collections['books'].retrieve()
    print("Collection already exists, retrieved:", collection)
```
